```python
import json
import os

def read_json(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return json.load(file)


def write_json(data, file_path):
    with open(file_path, 'w', encoding='utf-8') as file:
        json.dump(data, file, ensure_ascii=False, indent=4)

def merge_and_deduplicate(directory, file_pattern, max_number, output_file):
    all_data = []
    data_set = set()

    for i in range(2000, max_number + 1, 2000):
        file_name = os.path.join(directory, f"{file_pattern}{i}.json")
        if os.path.exists(file_name):
            data = read_json(file_name)
            for item in data:
                # Assuming each item can be converted to a JSON string as a unique identifier
                item_str = json.dumps(item, sort_keys=True)
                if item_str not in data_set:
                    data_set.add(item_str)
                    all_data.append(item)

    write_json(all_data, os.path.join(directory, output_file))

# 使用示例
merge_and_deduplicate('./data', 'test', 20000, 'merged_output.json')
```
